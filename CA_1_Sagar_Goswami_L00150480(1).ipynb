** Question 5**

** Stop words are the most general words in a vocabulory. These words needs to be filtered out before processing the text data as they have very less significance to the meaning of a sentence. A, the, an, it, and, about, I etc are some examples of stopwords. **

```python
#Creating a list of text tokens in the NLP document.
token_list = []
for token in doc_object:
    token_list.append(token.text)
```

```python
# Creating the list of stop words from the text tokens and printing few stop words
stop_words=[]
for word in token_list:
    lexeme = nlp.vocab[word]
    if lexeme.is_stop == True:
        stop_words.append(word) 
        
stop_words[:10]        
        
```

```python
removed_stopwords = token_list.remove('The')
```

** Question 7 **

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
# match for "stopword"
match1 = [{"LOWER": "woman"}]
# match for "stopwords"
match2 = [{"LOWER": "man"}]
# match for stop-word
match3 = [{"LOWER": "man"}, {"IS_PUNCT": True}, {"LOWER": "woman"}]
# match for stop-words
match4 = [{"LOWER": "woman"}, {"IS_PUNCT": True}, {"LOWER": "man"}]
```

```python
matcher.add("StopWord", None, match2, match1, match3, match4)
```

```python
def find_matches(doc_object):

    # find all matches within the doc object
    matches = matcher(doc_object)
    # For each item in the token_matches provide the following
    # match_id is the hash value of the identified token match
    men_count=0
    women_count=0
    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]
        matched_span = doc_object[start:end]      
        if (matched_span.text =='Man' or matched_span.text=='man'):
            men_count=men_count+1
        else:
            women_count=women_count+1
        print(f"{match_id:<{20}} {string_id:<{15}} {start:{5}} {end:{5}} {matched_span.text:{30}}")
    print("Total woman words occured for ",women_count)
    print("Total man words occured for",men_count)
    
```

```python
find_matches(doc_object)
```
